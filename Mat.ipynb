```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Graphs

```python
x = [1,3,4]
y = [1,3,4]

xr = [1,2,3]
yr = [1,4,9]

plt.plot(x,y, label='x', color='red', linewidth=2, marker='.',
         markeredgecolor='blue', markersize=10)
# all the things are self explanatory

# a short hand notation is used like [color][marker][line]
# 'g.--' this things is the short hand green,marker-'.',line --k
plt.plot(xr,yr, 'g.--', label='x**2', markeredgecolor='red', markersize=10) # this is about square plot
plt.savefig('myplot')
plt.legend()
plt.show()
```

```python
xnp = np.arange(0,5,0.5)

plt.figure(figsize=(5,3))
plt.plot(xnp, xnp**2, 'r.-', markeredgecolor='black', markersize=10, linewidth=2, label='x**2')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('TRIAL')
plt.legend()
plt.show()
```

### Bar Charts

```python
labels = ['A','B','C']
values = [2,4,3]

plt.figure(figsize=(5,3), dpi=100)

bars = plt.bar(labels, values)
bars[0].set_hatch('/')
bars[1].set_hatch('O')
bars[2].set_hatch('X')
plt.title('Bars')
plt.show()
```

## Fifa

```python
fifa = pd.read_csv(r"DataSets/fifa_data.csv", index_col=0)
fifa = fifa.drop(['Photo', 'Flag', 'Club Logo'], axis= 1)
fifa.head()
```

### Histogram

```python
bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins= bins)

plt.xticks(bins)
plt.xlabel('Skill levels')
plt.ylabel('Number of Players')
plt.show()
```

### Pie-Charts

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

plt.figure(figsize=(5,5), dpi=100)

plt.pie([left, right], labels=['left', 'right'], explode=[0.05, 0.05], autopct='% .2f %%')

plt.show()
```
