```python
doc = pd.read_csv("C:/Users/aayus/Dev/AI-ML/products-100.csv", index_col=0)
# doc = pd.DataFrame(doc)
doc.head()
```

```python
## Index based paradigms

doc.iloc[0]
# gives the 0th row
doc.iloc[:,0]
# gave 0th col not index col.
doc.iloc[:3,0]
# gives entry till 3rd row and for 0th col
```

```python
## Label-based Paradigm

# here we use loc which include [index, 'label_of_col]
doc.loc[:, "Name"]
# which is equal to below code
doc.iloc[0]
```

```python
## Conditional Selectional

# want to check whether the color is black or not for the given row
doc.Color=='Black'

# selecting those col where color is black
doc.loc[doc.Color == 'Black']
```
